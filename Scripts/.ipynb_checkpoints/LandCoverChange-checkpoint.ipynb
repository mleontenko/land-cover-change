```python
from osgeo import ogr, gdal, osr
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
#Paths
#Image-2019
path_B5_2019="../Images20190203/LC08_L1TP_029047_20190203_20190206_01_T1_B5.TIF"
path_B4_2019="../Images20190203/LC08_L1TP_029047_20190203_20190206_01_T1_B4.TIF"
#Image-2014
path_B5_2014="../Images20140205/LC08_L1TP_029047_20140205_20170307_01_T1_B5.TIF"
path_B4_2014="../Images20140205/LC08_L1TP_029047_20140205_20170307_01_T1_B4.TIF"

path_NDVI_2019 = 'RST_Output/NDVI2019.tif'
path_NDVI_2014 = 'RST_Output/NDVI2014.tif'

shp_clip = 'SHP_input/AOI.shp'

path_clipNDVI_2019 = 'RST_Output/clip_NDVI2019.tif'
path_clipNDVI_2014 = 'RST_Output/clip_NDVI2014.tif'

contours_clip2019 = 'SHP_output/ndvi_contour2019.shp'
contours_clip2014 = 'SHP_output/ndvi_contour2014.shp'
```

```python
B5_2019 = gdal.Open(path_B5_2019)
B4_2019 = gdal.Open(path_B4_2019)
B5_2014 = gdal.Open(path_B5_2014)
B4_2014 = gdal.Open(path_B4_2014)


B52019_Data = B5_2019.GetRasterBand(1).ReadAsArray().astype(np.float32)
B42019_Data = B4_2019.GetRasterBand(1).ReadAsArray().astype(np.float32)
B52014_Data = B5_2014.GetRasterBand(1).ReadAsArray().astype(np.float32)
B42014_Data = B4_2014.GetRasterBand(1).ReadAsArray().astype(np.float32)
```

```python
geotransform2019 = B5_2019.GetGeoTransform()

originX2019,pixelWidth2019,empty,finalY2019,empty2,pixelHeight2019=geotransform2019
cols2019 =  B5_2019.RasterXSize
rows2019 =  B5_2019.RasterYSize

geotransform2014 = B5_2014.GetGeoTransform()

originX2014,pixelWidth2014,empty,finalY2014,empty2,pixelHeight2014=geotransform2014
cols2014 =  B5_2014.RasterXSize
rows2014 =  B5_2014.RasterYSize
```

```python
finalX2019 = originX2019 + pixelWidth2019 * cols2019
originY2019 = finalY2019 + pixelHeight2019 * rows2019

finalX2014 = originX2014 + pixelWidth2014 * cols2014
originY2014 = finalY2014 + pixelHeight2014 * rows2014
```

```python
ndvi2019 = np.divide(B52019_Data - B42019_Data, B52019_Data+ B42019_Data,where=(B52019_Data - B42019_Data)!=0)
ndvi2019[ndvi2019 == 0] = -999

ndvi2014 = np.divide(B52014_Data - B42014_Data, B52014_Data+ B42014_Data,where=(B52014_Data - B42014_Data)!=0)
ndvi2014[ndvi2014 == 0] = -999
```

```python
Ds2019 = gdal.GetDriverByName('GTiff').Create(path_NDVI_2019, cols2019, rows2019,1,gdal.GDT_Float32)
Ds2019.SetProjection(B5_2019.GetProjection())
Ds2019.SetGeoTransform(geotransform2019)
Ds2019.GetRasterBand(1).WriteArray(ndvi2019)
Ds2019.GetRasterBand(1).SetNoDataValue(-999)
Ds2019 = None

Ds2014 = gdal.GetDriverByName('GTiff').Create(path_NDVI_2014, cols2014, rows2014,1,gdal.GDT_Float32)
Ds2014.SetProjection(B5_2014.GetProjection())
Ds2014.SetGeoTransform(geotransform2014)
Ds2014.GetRasterBand(1).WriteArray(ndvi2014)
Ds2014.GetRasterBand(1).SetNoDataValue(-999)
Ds2014 = None
```

```python
#plot image
NDVI2019 = gdal.Open(path_NDVI_2019)
ds2019 = NDVI2019.ReadAsArray()
plt.figure(figsize=(10,10))
im = plt.imshow(ds2019,vmin=-1, cmap='RdYlGn', extent=[originX2019,finalX2019,originY2019,finalY2019])#
plt.colorbar(im)
plt.title('NDVI 2019/02/03')
plt.xlabel('Este')
plt.ylabel('Norte')
plt.show()
```

```python
#plot image
NDVI2014 = gdal.Open(path_NDVI_2014)
ds2014 = NDVI2014.ReadAsArray()
plt.figure(figsize=(10,10))
im = plt.imshow(ds2014, vmin=0, cmap='RdYlGn', extent=[originX2014,finalX2014,originY2014,finalY2014])#
plt.colorbar(im)
plt.title('NDVI 2014/02/05')
plt.xlabel('Este')
plt.ylabel('Norte')
plt.show()
```

## Warp Rasters

```python
options = gdal.WarpOptions(cutlineDSName=shp_clip,cropToCutline=True)

OutTile2019 = gdal.Warp(destNameOrDestDS=path_clipNDVI_2019 ,srcDSOrSrcDSTab=path_NDVI_2019, options=options)
OutTile2019 = None

OutTile2014 = gdal.Warp(destNameOrDestDS=path_clipNDVI_2014 ,srcDSOrSrcDSTab=path_NDVI_2014, options=options)
OutTile2014 = None
```

```python
clip_NDVI2019 = gdal.Open(path_clipNDVI_2019)
originX2019clip,pixelWidth2019clip,empty,finalY2019clip,empty2,pixelHeight2019clip =clip_NDVI2019.GetGeoTransform()
cols2019clip =  clip_NDVI2019.RasterXSize
rows2019clip = clip_NDVI2019.RasterYSize

clip_NDVI2014 = gdal.Open(path_clipNDVI_2014)
originX2014clip,pixelWidth2014clip,empty,finalY2014clip,empty2,pixelHeight2014clip =clip_NDVI2014.GetGeoTransform()
cols2014clip =  clip_NDVI2014.RasterXSize
rows2014clip = clip_NDVI2014.RasterYSize

finalX2019clip = originX2019clip + pixelWidth2019clip * cols2019clip
originY2019clip = finalY2019clip + pixelHeight2019clip * rows2019clip

finalX2014clip = originX2014clip + pixelWidth2014clip * cols2014clip
originY2014clip = finalY2014clip + pixelHeight2014clip * rows2014clip
```

## Create Contourlines

```python
Dataset_ndvi = gdal.Open(path_clipNDVI_2019)#path_NDVI_2014
ndvi_raster = Dataset_ndvi.GetRasterBand(1)

ogr_ds = ogr.GetDriverByName("ESRI Shapefile").CreateDataSource(contours_clip2019)

prj=Dataset_ndvi.GetProjectionRef()#GetProjection()

srs = osr.SpatialReference(wkt=prj)#
#srs.ImportFromProj4(prj)

contour_shp = ogr_ds.CreateLayer('contour', srs)
field_defn = ogr.FieldDefn("ID", ogr.OFTInteger)
contour_shp.CreateField(field_defn)
field_defn = ogr.FieldDefn("elev", ogr.OFTReal)
contour_shp.CreateField(field_defn)
#Generate Contourlines
gdal.ContourGenerate(ndvi_raster, 0.1, 0, [], 1, -999, contour_shp, 0, 1)
ogr_ds = None
```

```python
Dataset_ndvi2014 = gdal.Open(path_clipNDVI_2019)
ndvi_raster2014 = Dataset_ndvi2014.GetRasterBand(1)

ogr_ds2014 = ogr.GetDriverByName("ESRI Shapefile").CreateDataSource(contours_clip2014)

prj2014=Dataset_ndvi2014.GetProjectionRef()#GetProjection()

srs = osr.SpatialReference(wkt=prj2014)#

contour_shp2014 = ogr_ds2014.CreateLayer('contour', srs)
field_defn = ogr.FieldDefn("ID", ogr.OFTInteger)
contour_shp2014.CreateField(field_defn)
field_defn = ogr.FieldDefn("elev", ogr.OFTReal)
contour_shp2014.CreateField(field_defn)
#Generate Contourlines
gdal.ContourGenerate(ndvi_raster2014, 0.1, 0, [], 1, -999, contour_shp2014, 0, 1)
ogr_ds2014 = None
```
